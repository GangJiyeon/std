```python
# 파이썬 자료형의 종류
a = 10
b = 3.14
c = "Hello, World!"
d = True
e = None
f = [1, 2, 3, 4, 5]
g = (1, 2, 3, 4, 5)
h = {'name': 'Alice', 'age': 30, 'city': 'New York'}
print(type(a))  # <class 'int'>
print(type(b))  # <class 'float'>
print(type(c))  # <class 'str'>
print(type(d))  # <class 'bool'>
print(type(e))  # <class 'NoneType'>
print(type(f))  # <class 'list'>
print(type(g))  # <class 'tuple'>
print(type(h))  # <class 'dict'>
```

```python
# 형변환
x = 10
y = 3.14
z = "100"
print(float(x))  # 10.0
print(int(y))    # 3
print(int(z))    # 100
print(str(x))    # '10'
print(str(y))    # '3.14'
print(str(z))    # '100'

print(str(x) + z)  # '10100'
```

```python
'''
물건 값의 7.5%가 부가세라고 하자. 물건값이 12345원일 때, 부가세를 소수점
2번째 자리까지 계산하는 프로그램
'''
price = 12345
tax_rate = 0.075

tax = price * tax_rate
print(f"부가세는 {tax:.2f}원입니다.")  # 부가세는 925.88원입니다.
```

```python
# 문제1) 식 (2)~(3) 중 식(1)과 계산 순서가 일치하는 코드 식은? (3)
x *= y + 1 # (1)
x = x * y + 1   # (2)
x = x * (y + 1) # (3)


#문제2) x=1000일 때, x += 1 과 x =+ 1 의 차이점은?
x = 1000
x += 1  # x = x + 1과 동일, x는 1001이 됨
print(x)  # 1001    
x = 1000
x =+ 1  # x = +1과 동일, x는 1이 됨
print(x)  # 1
```

```python
'''1626년에 아메리카 인디언들이 뉴욕의 맨하탄섬을 단돈 60길더(약 24달러)에
탐험가 Peter에게 팔았다고 한다. 382년 정도 경과한 맨하탄 땅값은 약 600억
달러라고 한다.
하지만 만약 인디언이 24달러를 은행의 정기예금에 입금 해두었다면 어떻게
되었을까? 예금 금리는 복리로 6%라고 가정하자. 그리고 382년이 지난 후에
는 원리금을 계산하여 보자.'''


initial_deposit = 24
annual_interest_rate = 0.06
years = 382
final_amount = initial_deposit * (1 + annual_interest_rate) ** years
print(f"382년 후의 원리금은 약 {final_amount:.2f} 달러입니다.")  # 382년 후의 원리금은 약 1.15
```

```python
# 표준입력
name = input("이름을 입력하세요: ")
print(name)  # 입력한 이름이 출력됨

# 표준출력
'''
print 함수의 속성
- sep: 출력할 값들 사이에 삽입되는 문자열을 지정 (기본값은 공백)
- end: 출력이 끝난 후에 삽입되는 문자열을 지정 (기본값은 줄바꿈)
'''

print("Hello", "World", sep="-")  # Hello-World
print("Hello", end=" ")  # Hello (줄바꿈 대신 공백)
print("World")  # World (이전 줄과 이어서 출력됨)   


'''
print 함수의 출력방법
1) 형식지정자 이용
2) format 메서드 이용
3) f-string 이용
'''

# 형식지정자 이용
name = "Alice"
age = 30
weight = 55.78652134
print("내 이름은 %s이고, 나이는 %d살이야" % (name, age))
print("내 몸무게는 %.3f kg이야" % weight)

# format 메서드 이용 - 중괄호에 인덱스 번호 안 적으면 순서대로 값이 들어감
print("내 이름은 {1}이고, 나이는 {0}살이야".format(age, name), end=" ")
print("10년 후에는 {}살이 돼".format(age + 10))

# f-string 이용
print(f"내 이름은 {name}이고, 나이는 {age}살이야")
print(f"10년 후에는 {age + 10}살이 돼")
```
